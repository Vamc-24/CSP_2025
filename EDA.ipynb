```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("tirupati_korlagunta_data.csv")
```

## Insight 1

```python
import seaborn as sns
sns.boxplot(data=df, y='pH').set_title('pH Distribution');
```



1. pH Levels
- Range: 6.82–8.16 (slightly alkaline)

- Compliance: All values within EPA standards (6.5–8.5).

- Insight: Stable, but minor fluctuations observed.

## Insight 2

```python
sns.scatterplot(data=df, x='EC (µS/cm)', y='TDS (mg/L)', hue='Hardness (mg/L)').set_title('EC vs. TDS vs. Hardness');
```

2. Electrical Conductivity (EC)
- Range: 313.77–1181.04 µS/cm.

- Insight: 17% of samples (5/30) exceed the ideal limit (800 µS/cm), indicating possible salinity/ion pollution.

- Correlation: Strong positive correlation with TDS (0.81) and hardness (0.65).

## Insight 3

```python
df.plot(x='Date', y='TDS (mg/L)', kind='line', title='Daily TDS Variation');
```

3. Total Dissolved Solids (TDS)
- Range: 173.98–756.73 mg/L.

- Compliance: 100% within EPA limits (< 500 mg/L).

- Insight: Inconsistent TDS despite stable pH; likely due to variable mineral runoff.

## Insight 4

```python
bins = [0, 150, 300, 450]  # Soft, Hard, Very Hard
df['Hardness Category'] = pd.cut(df['Hardness (mg/L)'], bins)
df['Hardness Category'].value_counts().plot(kind='bar');
```

4. Hardness
- Range: 152.76–445.99 mg/L (classified as "Hard" to "Very Hard").

- Insight: 63% of samples (19/30) exceed 200 mg/L, indicating high calcium/magnesium.

## Insight 5

```python
sns.heatmap(df[['Bacteria Count (CFU/mL)', 'Pesticide Level (µg/L)']].corr(), annot=True);
```

5. Bacteria Count
- Range: 4–144 CFU/mL.

- Red Flags: 37% of samples (11/30) exceed 100 CFU/mL (risk threshold).

- Pattern: Spikes correlate with low pesticide levels (e.g., Sample 140: high bacteria + low pesticides).

## Insight 6

```python
ax = df.plot(x='Date', y='Pesticide Level (µg/L)', color='red')
df.plot(x='Date', y='Bacteria Count (CFU/mL)', ax=ax, secondary_y=True, color='blue');
```

6. Pesticide Levels
- Range: 0.022–0.798 µg/L.

- Compliance: All within EPA limits (< 1 µg/L).

- Insight: Higher pesticides correlate with lower bacteria (*r* = -0.45), suggesting pesticides may suppress microbial growth.
